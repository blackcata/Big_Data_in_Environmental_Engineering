```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 

plt.rcParams['figure.dpi'] = 300
%config InlineBackend.figure_format = 'retina'
```

## Make input data using random integer

```python
data = np.random.randint(1,100,size=(5,2))
```

## Make tenforflow model 

Traing the neuron model with single layer.
Two 5  input data make the 5 output

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[2])])
model.summary()
```

## Calculate the output using tensorflow model

```python
output = model(data)
```

```python
print(output)
```

## Relationship between input & output

```python
fig, axes = plt.subplots(1,1,figsize=(5,5))

axes.plot(data[:,0],output,'o',color='tab:blue')
axes.plot(data[:,1],output,'o',color='tab:orange')

axes.set_xlabel("Input")
axes.set_ylabel("Output")
axes.grid(alpha=0.1)
```
